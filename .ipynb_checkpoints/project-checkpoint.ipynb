# Testing different machine learning algorithm

```python
target = data['TenYearCHD']
features = data.drop('TenYearCHD', axis=1)
# data scaling
scaler = MinMaxScaler()
scaled_feature = scaler.fit_transform(features)
# print(features)
```

```python
x = data.iloc[:,:-1]
y = data.TenYearCHD
x.head()
```

```python
# data spliting
X_train, X_test, y_train, y_test = train_test_split(scaled_feature,target, test_size=0.30, random_state=5)
```

```python
# Before oversampling data
print("Before oversampling, count of label '1' : {} ".format(sum(y_train == 1)))
print("Before oversampling, count of label '0' : {} ".format(sum(y_train == 0)))
```

```python
#  Using SMOTE technique for oversampling data
# Synthetic Minority Oversampling Technique
from imblearn.over_sampling import RandomOverSampler
smote = SMOTE(sampling_strategy=0.5, random_state=100)
X_train_sm, y_train_sm = smote.fit_resample(X_train,y_train.ravel() )

# After oversampling data
# counter = Counter(y_train)
print("After oversampling, count of label '1' : {} ".format(sum(y_train_sm == 1)))
print("After oversampling, count of label '0' : {} ".format(sum(y_train_sm == 0)))
```

```python
# X_train_sm, X_test, y_train_sm, y_test = train_test_split(x,y, test_size=0.3, random_state=10)
# X_train, X_test, y_train, y_test = train_test_split(scaled_feature,target, test_size=0.2, stratify= target)

st_x = StandardScaler()
X_train_sm = st_x.fit_transform(X_train_sm)
X_test = st_x.transform(X_test)
```

# KNN accuacy test

```python
classifier = KNeighborsClassifier(n_neighbors = 4)
classifier = classifier.fit(X_train_sm,y_train_sm)
knn_y_pred = classifier.predict(X_test)
#check accuracy
knn_accuracy = accuracy_score(y_test, knn_y_pred)
print('KNN Accuracy: {:.2f}'.format(knn_accuracy))
```

```python
test_score = accuracy_score(y_test, classifier.predict(X_test)) * 100
train_score = accuracy_score(y_train_sm, classifier.predict(X_train_sm)) * 100

results_df = pd.DataFrame(data=[["K-nearest neighbors", train_score, test_score]], 
                          columns=['Model', 'Training Accuracy %', 'Testing Accuracy %'])
results_df
```

## Confusion matrix of KNN 

```python
cm_knn = confusion_matrix(y_test, knn_y_pred)
print(cm_knn)
 
print('\n')
print("--------------------------")

# Confusion matrix can also be obtained using crosstab method of pandas.
pd.crosstab(y_test, knn_y_pred, rownames=['True'], colnames=['Predicted'], margins=True)
```

```python
f, ax =plt.subplots(figsize = (5,5))
sns.heatmap(cm_knn,annot = True, linewidths= 0.5, linecolor="red", fmt=".0f", ax=ax)
plt.xlabel("True label")
plt.ylabel("Predicted lable")
plt.show()
```

## ROC Curve for KNN

```python
# fpr,tpr,_=roc_curve(y_test,y_pred)
prediction = classifier.predict_proba(X_test)
fpr,tpr, threshold=roc_curve(y_test,prediction[:,1], pos_label=1)
roc_auc = auc(fpr,tpr)
# Plot of a ROC curve for a specific class
plt.figure()
plt.plot(fpr, tpr, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], 'k--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve of KNN')
plt.legend(loc="lower right")
plt.show()
```

```python
# printing the report
print(classification_report(y_test, knn_y_pred))
```


# Random Forest accuracy test

```python
randomForest = RandomForestClassifier()
randomForest.fit(X_train_sm, y_train_sm)
rf_y_pred = randomForest.predict(X_test)
rf_accuracy = accuracy_score(y_test, rf_y_pred)
print("Random Forest Classifier Accuracy: {:.2f}".format(rf_accuracy))
```

```python
test_score = accuracy_score(y_test, randomForest.predict(X_test)) * 100
train_score = accuracy_score(y_train_sm, randomForest.predict(X_train_sm)) * 100

results_df_2 = pd.DataFrame(data=[["Random Forest", train_score, test_score]], 
                          columns=['Model', 'Training Accuracy %', 'Testing Accuracy %'])
results_df = results_df.append(results_df_2, ignore_index=True)
results_df
```

## Confusion matrix of Random Forest

```python
# Confusion matrix
# Plot the confusion matrix in graph
cm_rf = confusion_matrix(y_test,rf_y_pred, labels=classifier.classes_)

# ploting with labels
disp = ConfusionMatrixDisplay(confusion_matrix=cm_rf, display_labels=classifier.classes_)
disp.plot()

# showing the matrix
plt.show()
```

```python
# printing the report
print(classification_report(y_test, rf_y_pred))

print('\n')
# Confusion matrix can also be obtained using crosstab method of pandas.
pd.crosstab(y_test, rf_y_pred, rownames=['True'], colnames=['Predicted'], margins=True)
```

## ROC curve of Random Forests

```python
 # fpr,tpr,_=roc_curve(y_test,y_pred)
prediction1 = randomForest.predict_proba(X_test)
fpr1,tpr1, threshold1=roc_curve(y_test,prediction1[:,1], pos_label=1)
roc_auc1 = auc(fpr1,tpr1)
# Plot of a ROC curve for a specific class
plt.figure()
plt.plot(fpr1, tpr1, label='ROC curve (area = %0.2f)' % roc_auc1)
plt.plot([0, 1], [0, 1], 'k--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve of Random Forests')
plt.legend(loc="lower right")
plt.show()
```
